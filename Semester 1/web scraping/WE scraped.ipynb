```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
url = "https://openlibrary.org/trending/daily"
response = requests.get(url)
html_content = response.text
soup = BeautifulSoup(html_content, "html.parser")
```

```python
h3s = soup.find_all("h3")
trending = []
for h3 in h3s:
    book_title = h3.get_text()
    result_details = h3.find_next(class_='resultDetails')
    author = h3.find_next(class_='bookauthor')
    author_name = author.get_text().strip() if author else 'Unknown'
    book_title = book_title.strip()
    year = result_details.get_text().strip() if result_details else 'Unknown'
    published_year = re.search(r'\b\d{4}\b', year)
    published_year = published_year.group(0) if published_year else 'Unknown'
    book_title = f"{book_title}"
    trending.append([book_title, author_name, published_year])
```

```python
trendingdf = pd.DataFrame(trending, columns=["Book Title", "Author Name", "Year Published"])
trendingdf
```
