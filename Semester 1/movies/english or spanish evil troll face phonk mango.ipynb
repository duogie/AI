```python
import pandas as pd
import seaborn as sns
```

```python
french = pd.read_json('french_movies.json')
english = pd.read_json('english_movies.json')
```

```python
sns.regplot(data=french, x="budget", y="revenue").set(title='French Movies')
```

```python
sns.regplot(data=english, x="budget", y="revenue").set(title='English Movies')
```
